```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Veri setini yükleme
data = pd.read_csv("C:/Users/gfbke/Desktop/Yapay_Zeka_Makine_Ogrenimi/Tahmin/Polynomial_Regresyon/Satis_Reklam.csv", encoding='latin-1')

# Reklam harcamaları ve satışlar verilerini alıyoruz
reklam_harcamalari = data["Reklam Harcamasi"].values.reshape(-1, 1)
satislar = data["Satis"].values

# Scatter plot çizimi
plt.figure(figsize=(10, 6))
plt.scatter(reklam_harcamalari, satislar, color='blue', label='Veri')

# Polinom regresyon modelini oluşturma
degree = 2  # Polinom derecesi
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(reklam_harcamalari)
model = LinearRegression()
model.fit(x_poly, satislar)

# Modelin tahmin ettiği değerleri hesaplama
x_values = np.linspace(min(reklam_harcamalari), max(reklam_harcamalari), 100).reshape(-1, 1)
x_poly_values = poly_features.transform(x_values)
y_values = model.predict(x_poly_values)

# Modelin doğrusunu çizme
plt.plot(x_values, y_values, color='red', label=f'Polinom Derecesi {degree}')
plt.title('Reklam Harcamaları ve Satışlar')
plt.xlabel('Reklam Harcamaları')
plt.ylabel('Satışlar')
plt.grid(True)
plt.legend()
plt.show()
```
