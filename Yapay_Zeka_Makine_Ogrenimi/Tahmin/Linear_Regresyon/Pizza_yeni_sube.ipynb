```python
# Kütüphaneleri içe aktaralım
import pandas as pd
import matplotlib.pyplot as plt

# Verileri Excel dosyasından okuyalım
data = pd.read_excel("C:/Users/gfbke/Desktop/Yapay_Zeka_Makine_Ogrenimi/Tahmin/Linear_Regresyon/Pizza_yeni_sube.xlsx", sheet_name="Sayfa1")

# Verileri değişkenlere atayalım
kira_bedeli = data["Kira Bedeli"].values  # Sütun adına göre veri çekme ve numpy dizisine dönüştürme
pizza_fiyati = data["Pizza Fiyatı"].values  # Sütun adına göre veri çekme ve numpy dizisine dönüştürme

# Lineer regresyon modelini oluşturalım
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(kira_bedeli.reshape(-1, 1), pizza_fiyati)

# Eğimi ve kesme noktasını yazdıralım
print("Eğim:", model.coef_[0])
print("Kesme noktası:", model.intercept_)

# Tahmini pizza fiyatını hesaplayalım
yeni_kira_bedeli = 8000
tahmini_pizza_fiyati = model.predict([[yeni_kira_bedeli]])

# Verileri ve tahmini görselleştirelim
plt.scatter(kira_bedeli, pizza_fiyati, color="blue", marker="x")
plt.plot(kira_bedeli, model.predict(kira_bedeli.reshape(-1, 1)), color="red", linewidth=2)
plt.xlabel("Kira Bedeli", fontsize=12)
plt.ylabel("Pizza Fiyatı", fontsize=12)
plt.grid(True)
plt.show()

print("Yeni şube için tahmini pizza fiyatı:", tahmini_pizza_fiyati)
```
