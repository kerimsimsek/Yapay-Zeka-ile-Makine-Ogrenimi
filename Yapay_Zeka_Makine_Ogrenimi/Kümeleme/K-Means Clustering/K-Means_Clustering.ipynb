```python
import pandas as pd
import numpy as np

# Rastgele veri oluşturma
np.random.seed(0)
n_samples = 1000

# Gen ifadesi ve protein etkileşimi özelliklerini oluşturma
gen_ifadesi = np.random.uniform(low=0, high=100, size=n_samples)
protein_etkileşimi = np.random.uniform(low=0, high=100, size=n_samples)

# DataFrame oluşturma
df = pd.DataFrame({
    'Gen_Ifadesi': gen_ifadesi,
    'Protein_Etkileşimi': protein_etkileşimi
})

# Veri setini CSV dosyasına yazdıralım
df.to_csv('gen_veri_seti.csv', index=False)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Gen veya protein özelliklerini içeren veri setini oluşturun
data = pd.read_csv("C:/Users/gfbke/OneDrive/Masaüstü/kp/Kümeleme/K-Means Clustering/gen_veri_seti.csv")

# Bağımsız değişkenleri seçin (örneğin, gen ifadesi seviyeleri, protein etkileşimleri)
X = data[["Gen_Ifadesi", "Protein_Etkileşimi"]]

# K-Means modelini oluşturun
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Küme merkezlerini görselleştirin
plt.scatter(X["Gen_Ifadesi"], X["Protein_Etkileşimi"], c=kmeans.labels_, cmap="rainbow")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,  1], color="black", marker="x", s=100)
plt.xlabel("Gen Ifadesi")
plt.ylabel("Protein Etkileşimi")
plt.title("K-Means Clustering")
plt.show()
```
